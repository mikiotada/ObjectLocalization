```python
def Img2bbox(path2gray_img, path2original_img, path2bbox):
    ImgID2bbox = {}
    for img_file in os.listdir(path2gray_img):
        imgID = img_file.split(".")[0]
        gray_activation_map_arr = cv2.imread(path2gray_img+"/"+img_file)
        
        gray = cv2.cvtColor(gray_activation_map_arr, cv2.COLOR_BGR2GRAY)
        thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)[1]
        cnts = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
        cnts = cnts[0] if len(cnts) == 2 else cnts[1]
        
        
        
        bbox = []
        for c in cnts:
            x, y, w, h = cv2.boundingRect(c)
            bbox.append([x, y, w, h])
#             cv2.rectangle(gray_activation_map_arr, (x, y), (x + w, y + h), color, thickness)
        
        ImgID2bbox[imgID] = bbox
    return ImgID2bbox
#         cv2.imwrite(path2bbox+imgID+".jpg", gray_activation_map_arr)
```

```python
def drawbbox(ImgID2bbox, path2original_img, path2bbox):
    color = (36,255,12)
    thickness = 2
    for imgID in ImgID2bbox:
        imgfile = path2original_img+imgID+".jpg"
        print(imgfile)
        original_img_arr = cv2.imread(imgfile)
        resize_original_img_arr = cv2.resize(original_img_arr, (256, 256))

        for bbox_info in ImgID2bbox[imgID]:
            x = bbox_info[0]
            y = bbox_info[1]
            w = bbox_info[2]
            h = bbox_info[3]
            cv2.rectangle(resize_original_img_arr, (x, y), (x + w, y + h), color, thickness)
        print(resize_original_img_arr.shape)
        cv2.imwrite(path2bbox+imgID+".jpg", resize_original_img_arr)
```
