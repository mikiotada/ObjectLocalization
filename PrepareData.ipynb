```python
import os
import numpy as np
import pandas as pd
from PIL import Image
import cv2 
```

## Resive all the imiages to 570*530

```python
resize = (570, 530)
def resizeImages(path2img, path2resize_img):
    for img_file in os.listdir(path2img):
#         print(img_file)
        original_img_arr = cv2.imread(path2img+img_file)
        resize_original_img_arr = cv2.resize(original_img_arr, resize)
        cv2.imwrite(path2resize_img+img_file, resize_original_img_arr)
```

```python
def img2class(path2img, label):
    images = [file for file in os.listdir(path2img) if file.split(".")[1] == "jpg"]
    size = len(images)
    labels = [label]*size
    data = {'images':images, 'labels':labels}
    df = pd.DataFrame(data)
    return df
    
```
